# 0. Подготовительный этап

```python
%matplotlib inline
```

```python
from scipy.misc import imread
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
import scipy.ndimage as ndi
import skimage.filters as skmg_filt
import skimage.morphology as skmg_morph
from skimage import feature
```

```python
def rgb2gray(rgb):
    """
    Функция переводит 3х-канальное rgb изображение в 
    одноканальное серошкальное
    """
    return np.dot(rgb[...,:3],[0.299, 0.587, 0.114])
```

```python
im_blood = rgb2gray(imread('blood.jpg',False, 'RGB' ))
plt.imshow(im_blood, cmap='gray')
plt.axis('off')
plt.show()
```

## Вспомогательная функция

```python
def plot_comparison(original, filtered_1, filter_name_1, filtered_2, filter_name_2, filtered_3, filter_name_3):

    fig, ax = plt.subplots(2,2, figsize = (15,15))

    ax[0,0].imshow(original, cmap=plt.cm.gray)
    ax[0,0].set_title('Исходное изображение')
    ax[0,0].axis('off')

    ax[0,1].imshow(filtered_1, cmap=plt.cm.gray)
    ax[0,1].set_title(filter_name_1)
    ax[0,1].axis('off')

    ax[1,0].imshow(filtered_2, cmap=plt.cm.gray)
    ax[1,0].set_title(filter_name_2)
    ax[1,0].axis('off')

    ax[1,1].imshow(filtered_3, cmap=plt.cm.gray)
    ax[1,1].set_title(filter_name_3)
    ax[1,1].axis('off')
```

# 1) Бинаризация изображения

```python
fig, ax = skmg_filt.try_all_threshold(im_blood, figsize=(10,8), verbose=False)
print('Разные подходы к бинаризации')
plt.show()
```

```python
thresh_min = skmg_filt.threshold_mean(im_blood)
im_blood_bin = im_blood > thresh_min
fig, ax = plt.subplots(2,2, figsize = (10,10))

ax[0,0].imshow(im_blood, cmap=plt.cm.gray)
ax[0,0].set_title('Исходное изображение')

ax[0,1].hist(im_blood.ravel(), bins=256)
ax[0,1].set_title('Гистограмма')

ax[1,0].imshow(im_blood_bin, cmap=plt.cm.gray)
ax[1,0].set_title('Бинаризованное изображение')

ax[1,1].hist(im_blood.ravel(), bins=256)
ax[1,1].axvline(thresh_min, color='r')
plt.show()
```

### Наиболее подходящий вариант бинаризации в случае использовании порога threshold_mean

# 2) Генерации структурных элементов

```python
plt.imshow(skmg_morph.diamond(5), cmap = 'gray')

def struct_diamond(n):
    return skmg_morph.diamond(n)
```

```python
plt.imshow(skmg_morph.disk(6), cmap = 'gray')

def struct_disk(n):
    return skmg_morph.disk(n)
```

```python
plt.imshow(skmg_morph.square(6), cmap = 'gray')

def struct_square(n):
    return skmg_morph.square(n)
```

## Примечание: поскольку гарницы элементов изображения черные, а фон белый, все морфологические операции выглядят как инвертированые, т.е. при дилатации черные элементы будут истончаться, поскольку расширяется белая область, а при эрозии будут наращиваться границы черных элементов. Таким образом, условно можно сказать, что морфологические операции проводятся над фоном (дилатация, эрозия фона)

# 3) Дилатация

```python
im_blood_dilat_dia_1 = ndi.binary_dilation(im_blood_bin, structure = struct_diamond(1), iterations = 1)
im_blood_dilat_dia_2 = ndi.binary_dilation(im_blood_bin, structure = struct_diamond(2), iterations = 1)
im_blood_dilat_dia_3 = ndi.binary_dilation(im_blood_bin, structure = struct_diamond(3), iterations = 1)

plot_comparison(im_blood_bin, 
                im_blood_dilat_dia_1, 'После дилатации/diamond_1', 
                im_blood_dilat_dia_2, 'После дилатации/diamond_2', 
                im_blood_dilat_dia_3, 'После дилатации/diamond_3')
```

### Выбираем diamond с парметром 1

```python
im_blood_dilat_dis_1 = ndi.binary_dilation(im_blood_bin, structure=struct_disk(1), iterations = 1)
im_blood_dilat_dis_2 = ndi.binary_dilation(im_blood_bin, structure=struct_disk(2), iterations = 1)
im_blood_dilat_dis_3 = ndi.binary_dilation(im_blood_bin, structure=struct_disk(3), iterations = 1)

plot_comparison(im_blood_bin, 
                im_blood_dilat_dis_1, 'После дилатации/disk_1', 
                im_blood_dilat_dis_2, 'После дилатации/disk_2', 
                im_blood_dilat_dis_3, 'После дилатации/disk_3')
```

### Выбираем disk с параметром 1

```python
im_blood_dilat_squ_1 = ndi.binary_dilation(im_blood_bin, structure=struct_square(1), iterations = 1)
im_blood_dilat_squ_2 = ndi.binary_dilation(im_blood_bin, structure=struct_square(2), iterations = 1)
im_blood_dilat_squ_3 = ndi.binary_dilation(im_blood_bin, structure=struct_square(3), iterations = 1)

plot_comparison(im_blood_bin, 
                im_blood_dilat_squ_1, 'После дилатации/square_1', 
                im_blood_dilat_squ_2, 'После дилатации/square_2', 
                im_blood_dilat_squ_3, 'После дилатации/square_3')
```

### Выбираем square с параметром 2

# Визуализируем отобранные варианты

```python
plot_comparison(im_blood_bin, 
                im_blood_dilat_dia_1, 'После дилатации/diamond_1', 
                im_blood_dilat_dis_1, 'После дилатации/disk_1', 
                im_blood_dilat_squ_2, 'После дилатации/square_2')
```

# Выделение границ

### Воспользуемся фильтром Кэнни

### Применим к исходному изображению 

```python
edges_init_1 = feature.canny(im_blood_bin, sigma=0.1)
edges_init_2 = feature.canny(im_blood_bin, sigma=0.2)
edges_init_3 = feature.canny(im_blood_bin, sigma=0.5)

plot_comparison(im_blood_bin, 
                edges_init_1, 'Фильтр Кэнни/sigma = 0.1', 
                edges_init_2, 'Фильтр Кэнни/sigma = 0.2', 
                edges_init_3, 'Фильтр Кэнни/sigma = 0.3')
```

### Наилучший вариант edges_init_1 с sigma = 0.1 

### Применим к изображению после размыкания


```python
edges_closing_1 = feature.canny(im_blood_closing_squ_2, sigma=0.1)
edges_closing_2 = feature.canny(im_blood_closing_squ_2, sigma=0.2)
edges_closing_3 = feature.canny(im_blood_closing_squ_2, sigma=0.5)

plot_comparison(im_blood_bin, 
                edges_closing_1, 'Фильтр Кэнни/sigma = 0.1', 
                edges_closing_2, 'Фильтр Кэнни/sigma = 0.2', 
                edges_closing_3, 'Фильтр Кэнни/sigma = 0.3')
```

### Наилучший вариант edges_closing_1 с sigma = 0.1 

### Применим к изображению после дилатации

```python
im_blood_dilat_dia_1
edges_dilat_1 = feature.canny(im_blood_dilat_dia_1, sigma=0.1)
edges_dilat_2 = feature.canny(im_blood_dilat_dia_1, sigma=0.2)
edges_dilat_3 = feature.canny(im_blood_dilat_dia_1, sigma=0.5)

plot_comparison(im_blood_bin, 
                edges_dilat_1, 'Фильтр Кэнни/sigma = 0.1', 
                edges_dilat_2, 'Фильтр Кэнни/sigma = 0.2', 
                edges_dilat_3, 'Фильтр Кэнни/sigma = 0.3')
```

### Наилучший вариант edges_dilat_2 с sigma = 0.2

### Сравним лучшие варианты

```python
plot_comparison(im_blood_bin, 
                edges_init_1, 'Фильтр Кэнни на исходном изображении/sigma = 0.1', 
                edges_closing_1, 'Фильтр Кэнни после размыкания/sigma = 0.1', 
                edges_dilat_2, 'Фильтр Кэнни после дилатации/sigma = 0.2')
```

### В качестве наилучшего варианта выделения границ выберем применение фильтра Кэнни к изображению после дилатации. Метод Кэнни выбран среди прочих алгоритмов выделения границ (оператор Собеля, Робертса и пр.), поскольку он признан оптимальным по критериям выделения, локализации и минимизации нескольких откликов одного края.

```python
plt.imshow(edges_dilat_2, cmap = 'gray')
plt.axis('off')
plt.title('Наилучший вариант выделения границ: Метод Кэнни после дилатации')
plt.show()
```
