## 1. Построение маски 2D фильтра Габора в каноническом виде при различных значениях параметров 

Импортируем библиотеки

```python
from cmath import pi, cos, sin, exp, sqrt
import matplotlib.pyplot as plt
from scipy.misc import imread
import scipy.ndimage as ndi
import numpy as np
import copy
```

Зададим параметры комплексной синусоиды:  
S(x,y)=e^(i(2π(u_0 x+v_0 y)+P))  

```python
u0 = 0.1
v0 = 0.1
P = 0
```

где u_0,v_0 – пространственные частоты синусоиды, P - фаза синусоиды

Зададим параметры двумерной функции Гаусса:   
G(x,y)=e^(-π[a^2 (((x-x_0 )cosθ+(y-y_0 )sinθ))^2+b^2 ((-(x-x_0 )sinθ+(y-y_0 )cosθ))^2])

```python
a = 0.02
b = 0.03

theta=pi/4

M = 100
N = 100

x0 = M/2
y0 = N/2
```

где x_0=M/2,y_0=N/2 - центр окна Гаусса, M×N – размер изображения, a,b - параметры формы окна, θ (theta) - угол поворота окна

Зададим параметры 2D функции Габора:  
    Gabor(x,y)=K∙S(x,y)∙G(x,y)

```python
K = 1
```

K - амплитудный множитель

Создадим пустые двумерные numpy-массивы для хранения значений функций

```python
S = np.empty((100,100), dtype=complex)
G = np.empty((100,100), dtype=complex)
Gabor = np.empty((100,100), dtype=complex)
```

Во вложенном цикле рассчитаем значения комплексной синусоиды, двумерной функции Гаусса и двумерной функции Габора

```python
for x in range(0,100):
    for y in range(0, 100):
        S[x,y] = exp(1j*(2*pi*(u0*x + v0*y)))
        G[x,y] = exp(-pi*((a**2)*((x - x0)*cos(theta) + (y - y0)*sin(theta))**2 + \
                         (b**2)*(-(x - x0)*sin(theta) + (y - y0)*cos(theta))**2))
        Gabor[x,y] = K*S[x,y]*G[x,y]
```

Визуализируем полученный результат

```python
plt.imshow(np.real(S), cmap="gray")
plt.axis("off")
plt.title("Рисунок 1 - Действительная часть 2D комплексной синусоиды")
plt.show()
```

```python
plt.imshow(np.real(G), cmap="gray")
plt.axis("off")
plt.title("Рисунок 2 - 2D функция Гаусса ")
plt.show()
```

```python
plt.imshow(np.real(Gabor), cmap="gray")
plt.axis("off")
plt.title("Рисунок 3 -Действительная часть 2D функции Габора")
plt.show()
```

## 2. Формирование Фурье-образов 2D фильтра Габора в каноническом виде 

Применим двумерное БПФ к двумерной функции Габора и сместим нулевую частоту спектра в центре

```python
Gabor_fft = np.fft.fftshift(np.fft.fft2(Gabor))
```

```python
plt.imshow(np.real(Gabor_fft), cmap="gray")
plt.axis("off")
plt.title("Рисунок 4 - Фурье-образ 2D фильтра Габора")
plt.show()
```

## 3. Применение полученных фильтров для обработки тестового изображения 

```python
def rgb2gray(rgb):
    """
    Функция переводит 3х-канальное rgb изображение в 
    одноканальное серошкальное
    """
    return np.dot(rgb[...,:3],[0.299, 0.587, 0.114])
```

Считаем тестовое изображение

```python
im_test = rgb2gray(imread('Images/test.png',False ,'RGB'))
plt.imshow(im_test, cmap='gray')
plt.title("Рисунок 5 - Тестовое изображение")
plt.axis('off')
plt.show()
```

Применем фильтр Габора к тестовому изображению

```python
im_filtered = ndi.convolve(im_test, np.real(Gabor), mode='nearest')
```

```python
plt.imshow(im_filtered, cmap="gray")
plt.title("Рисунок 6 - Результат фильтрации исходного изображения")
plt.axis('off')
plt.show()
```

## 4. Построение банка 2D фильтров Габора 

Зададим параметры базисного фильтра Габора:  
g(x,y,μ,ν)=k_ν^2∙1/σ^2 ∙e^(-(k_ν^2 (x^2+y^2 ))/(2σ^2 ))∙[e^(jk_ν (xcosω_μ+ysinω_μ))-e^(-σ^2/2) ];   
k_ν=k_max/f^ν;  
k_max=π/2;   
f=√(2);   
ω_μ=πμ/8;    
μ=0,1,2,3,4,5,6,7;    
ν=0,1,2,3,4;  
σ=2π

```python
k_max = pi/2;
f = sqrt(2);
sigma = 2*pi;
```

Создадим пустые двумерные numpy-массивы:  
- Gabor_bank для хранения фильтров банка Габора (object-type элементы) и 
- g для вычисления фильтров Габора (complex-type элементы)

```python
Gabor_bank = np.empty((5,8), dtype=np.object)
g = np.empty((100,100), dtype=complex)
```

Рассчитаем в цикле банк фильтров Габора 

```python
for nu in range(0,5):
    for mu in range(0,8):
        omega = (pi*mu)/8
        k_nu = k_max/(f**nu)
        
        for x in range(0,100):
            for y in range(0,100):
                g[x,y] = (k_nu**2)*(1/sigma**2)*exp(-(k_nu**2*((x-x0)**2 + \
                         (y-y0)**2))/(2*sigma**2))*(exp(1j*k_nu*((x-x0)*cos(omega) +\
                         (y-y0)*sin(omega))) - exp(-(sigma**2/2)))
        Gabor_bank[nu,mu] = copy.copy(g)       
```

Визуализируем банк фильтров Габора

```python
fig, axes = plt.subplots(5, 8, figsize=(15,10))
fig.suptitle("Рисунок 7 - Банк фильтров Габора")
for i in range(0,5):
    for j in range(0,8):
        axes[i,j].imshow(np.real(Gabor_bank[i,j]), cmap="gray")
        axes[i,j].axis("off")        
plt.show()
```

## 5. Формирование Фурье-образов банка 2D фильтров Габора 

Визуализация Фурье-образов банка фильтров Габора

```python
fig, axes = plt.subplots(5, 8, figsize=(25,15))
fig.suptitle("Рисунок 8 - Фурье-образы банка фильтров Габора")
for i in range(0,5):
    for j in range(0,8):
        axes[i,j].imshow(np.real(np.fft.fftshift(np.fft.fft2(Gabor_bank[i,j]))), cmap="gray")
        axes[i,j].axis("off")
plt.show()
```

```python
fig, axes = plt.subplots(5, 8, figsize=(25,15))
fig.suptitle("Рисунок 9 - Тестовое изображение после обработки фильтрами")
for i in range(0,5):
    for j in range(0,8):
        axes[i,j].imshow(ndi.convolve(im_test, np.real(Gabor_bank[i,j]), mode='nearest'), cmap="gray")
        axes[i,j].axis("off") 
        axes[i,j].set_title(f"[{i},{j}]")
plt.show()
```

## 7. Выводы о влиянии параметров фильтра Габора на результаты преобразования тестового изображения

Из рисунка 9 видно, что для данного тестового изображения (лица) границы объекта распознаются точнее в случае nu = 0, где k_nu = k_max/(f^nu). В случае увеличения nu (снижения частоты) фильтр также определяет расположение характерных объектов, например, губы и глаза на скерии изображений [0:4,0], но с ростом nu точность локализации снижается.  
Также необходимо отметить, что различные значения параметра mu в omega = (pi*mu)/8 (ориентация) позволяют определять отдельные характерные точки изображения. Например:
- при mu = 0 подчеркивается положение глаз и губ;
- при mu = 4 подчеркиваются вертикальные границы лица;
- при mu = 5 (особенно на рисунке [1,5]) подчеркиваются общие черты лица.
